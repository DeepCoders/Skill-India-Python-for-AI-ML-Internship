## Read Data from JSON

```python
import json

fd = open("record.json",'r')
data = fd.read()
fd.close()

products = json.loads(data)
```

```python
products
```

## Add New Item into Inventory

```python
prod_id = str(input("Enter product id: "))
name = str(input("Enter product name: "))
price = int(input("Enter price: "))
quantity = int(input("Enter quantity: "))
brand = str(input("Enter product brand: "))
category = str(input("Enter product category: "))

products[prod_id] = {'name': name, 'price': price, 'quantity': quantity, 'brand': brand, 'category': category}

new_data = json.dumps(products)

fd = open("record.json",'w')
fd.write(new_data)
fd.close()
```

```python
products
```

## Update Stocks

```python
prod_id = str(input("Enter product id: "))
stocks = int(input("Enter the stocks you want to add: "))
products[prod_id]['quantity'] += stocks

updated_data = json.dumps(products)

fd = open("record.json",'w')
fd.write(updated_data)
fd.close()
```
